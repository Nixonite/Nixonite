So mergesort and quicksort both have O(nlogn) compexity, but Insertionsort runs on O(n^2). How can we see this in action?

```python
import numpy as np
import time
import matplotlib.pyplot as plt
%matplotlib inline
```

Below is some code to generate a random sample of integers with a minimum value of 1 and a maximum value of 1000000000. Here's a test sample.

```python
randomList = np.random.random_integers(1,1000000000,size=1000000)
print len(randomList)
print randomList
```

The code below shows three lists. One for storing the computation times of sorting for Quicksort, another for Mergesort, and one for Insertionsort. By sorting the same list of random integers, we can compare the times and plot them to see how they match up. We should expect to see the same <b>slope</b> for the computational times of MergeSort and QuickSort since they both [almost always] have O(nlogn) complexity.

```python
def sort_insertion(my_list):
    #taken from http://runnable.com/Uq_i4quWT8lOAAAv/algorithms-insertion-sort-python
    for i in range(1,len(my_list)):
        val_current = my_list[i]
        pos = i 
        # check backwards through sorted list for proper pos of val_current
        while((pos > 0) and (my_list[pos-1] > val_current)):
            my_list[pos] = my_list[pos-1]
            pos = pos-1       
        if pos != i:
            my_list[pos] = val_current 
    return my_list 

quickSortTimes = []
mergeSortTimes = []
insertionSortTimes = []
sizes = [10**i for i in range(1,5)]
for i in sizes:
    randomList = np.random.random_integers(1,1000000000,size=i)
    
    startTime = time.time()
    quickList = np.sort(randomList,kind='quicksort')
    runTime = time.time()-startTime
    quickSortTimes.append(runTime)
    
    startTime = time.time()
    mergeList = np.sort(randomList,kind='mergesort')
    runTime = time.time()-startTime
    mergeSortTimes.append(runTime)
    
    startTime = time.time()
    bubbleList = sort_insertion(randomList)
    runTime = time.time()-startTime
    insertionSortTimes.append(runTime)
```

Now let's plot the values to see the results...

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(sizes,quickSortTimes,marker='x',c='b',label='Quicksort')
plt.plot(sizes,mergeSortTimes,marker='x',c='r',label='Mergesort')
plt.plot(sizes,insertionSortTimes,marker='x',c='g',label='Insertionsort')
plt.xscale('log')
plt.yscale('log')
plt.xlabel("size of unsorted list - log scale")
plt.ylabel("seconds of computation - log scale")
plt.legend(loc=2)
plt.grid()
plt.title("Quicksort vs Mergesort vs Insertionsort")

plt.subplot(1,2,2)
plt.plot(sizes,quickSortTimes,marker='x',c='b',label='Quicksort')
plt.plot(sizes,mergeSortTimes,marker='x',c='r',label='Mergesort')
plt.plot(sizes,insertionSortTimes,marker='x',c='g',label='Insertionsort')
plt.xlabel("size of unsorted list")
plt.ylabel("seconds of computation")
plt.legend(loc=2)
plt.grid()
plt.title("Quicksort vs Mergesort vs Insertionsort")

plt.show()
```

So you may have noticed that the left plot is log-scaled. Why did I do this? Let's assume that the runtime can be denoted as

$$t=n^2, \text{ where } n^2 \text{ is the complexity of an algorithm.}$$

This also happens to be the runtime as Insertionsort. Then if we were to take the log of both sides...

$$\log{t} = 2\log{n}$$

Aha, so if we plot this log-log scale, we should expect a graph with slope 2. So it is indeed, if we see the green line in the left plot, it has a slope ~2 whereas the quicksort and mergesort rightfully have a slope of about 1. Why 1?

The complexity of Quicksort (average case) and Mergesort is O(nlogn)...

$$
t = n\log{n}
\\
\log{t} = \log{n} + \log{\log{n}}
$$

Since complexity really only matters for the highest order (because they far surpass the speed of lower order values over time), we see here that

$$
\log{t} = \log{n}
$$

Which means we should expect a linear slope for large values, as we can see between the last two xticks in the first plot for the red and blue lines.
